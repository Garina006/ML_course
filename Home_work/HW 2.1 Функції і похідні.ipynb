### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
x = np.linspace(0, 10, 50)
f_x = 10 * x - 5
df = 10 #для даної функції похідна є константою (df = df10x - df5 = 10 * 1 * x ** (1-1) - 0 = 10 *1 * 1 - 0 = 10)
df_vals = [df] * len(x)
print(f"Похідні для кожної х: {df_vals}")
plt.figure(figsize=(10, 5))
plt.plot(x, f_x, label="f(x) = 10x - 5", color='blue')
plt.plot(x, df_vals, label=f"f'(x) = {df}", color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Графік функції f(x) та її похідної f'(x)")
plt.legend()
plt.grid()
plt.show()
```

**Висновок:**  
Збільшення кількості годин навчання лінійно впливає на оцінку: чим більше годин, тим вище оцінка.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



Розрахуємо похідну для функції $ f(x) = \sin(x) $ та зобразимо обидві функції на графіку

```python
x = sp.Symbol('x')
f_x = sp.sin(x)

derivative_f = sp.diff(f_x, x)

f_deriv_func = sp.lambdify(x, derivative_f)

xs = np.linspace(0, 10, 100)
f_func = sp.lambdify(x, f_x)
f_x_vals = f_func(xs)
df_vals = [f_deriv_func(x) for x in xs]
print(f"Похідні: {df_vals}")

plt.figure(figsize=(10, 5))
plt.plot(xs, f_x_vals, label="f(x) = sin(x)", color='blue')
plt.plot(xs, df_vals, label=f"f'(x) = cos(x)", color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Графік функції f(x) та її похідної f'(x)")
plt.legend()
plt.grid()
plt.show()
```

Розрахуємо похідну для функції $ f(x) = \ln(x) $ та зобразимо обидві функції на графіку

```python
x = sp.Symbol('x')
f_x = sp.ln(x)

derivative_f = sp.diff(f_x, x)

f_deriv_func = sp.lambdify(x, derivative_f)

xs = np.linspace(1, 10, 100)
f_func = sp.lambdify(x, f_x)
f_x_vals = f_func(xs)
df_vals = [f_deriv_func(x) for x in xs]
print(f"Похідні: {df_vals}")

plt.figure(figsize=(10, 5))
plt.plot(xs, f_x_vals, label="f(x) = ln(x)", color='blue')
plt.plot(xs, df_vals, label=f"f'(x) = 1/x", color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Графік функції f(x) та її похідної f'(x)")
plt.legend()
plt.grid()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.Symbol('x')
f_x = x ** 2 - 5 * x + 1

derivative_f = sp.diff(f_x, x)

f_deriv_func = sp.lambdify(x, derivative_f)

xs = np.linspace(0, 100, 100)
f_x_vals = xs ** 2 - 5 * xs + 1
df_vals = [f_deriv_func(x) for x in xs]

plt.figure(figsize=(10, 5))
plt.plot(xs, f_x_vals, label="f(x) = x**2 - 5x + 1", color='blue')
plt.plot(xs, df_vals, label=f"f'(x) = 2x - 5", color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Графік функції f(x) та її похідної f'(x)")
plt.legend()
plt.grid()
plt.show()
```
